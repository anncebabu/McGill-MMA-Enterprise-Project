```python
data['order date (DateOrders)']=pd.to_datetime(data['order date (DateOrders)'])
```

```python
data=data.rename(columns={'order date (DateOrders)':'order_date'})
```

```python
data['Year']=data['order_date'].dt.year
```

```python
data['Month']=data['order_date'].dt.month
```

```python
data.sort_values(by='order_date', inplace=True)
```

```python
data.drop(columns=['order_date'],inplace=True)
```

```python
# Aggregate order quantities by region
grouped_data = data.groupby(['Year', 'Market', 'Month', 'Order Country',
                                        'Order Region', 'Order State', 'Product Name', 
                                         'Category Name','Customer Segment'])
```

```python
grouped_data.head()
```

```python
supplyChain_group1 = grouped_data["Order Item Quantity"].sum().reset_index()
supplyChain_group2 = grouped_data["Sales"].sum().reset_index()
supplyChain_group3 = grouped_data["Days for shipping (real)"].mean().reset_index()
```

```python
supplyChain_group4 = grouped_data["Product Price"].mean().reset_index()
supplyChain_group5 = grouped_data["Product Status"].mean().reset_index()
```

```python
supplyChain_df = supplyChain_group1.copy()
supplyChain_df["Sales"] = supplyChain_group2["Sales"]
supplyChain_df["Days for shipping (real)"] = supplyChain_group3["Days for shipping (real)"]
supplyChain_df["Product Price"] = supplyChain_group4["Product Price"]
supplyChain_df["Product Status"] = supplyChain_group5["Product Status"]
```

```python
cat_col=[]
for col in data.columns:
    if data[col].dtype =='object':
        cat_col.append(col)
```

```python
# Dummy variables
supplyChain_df = pd.get_dummies(supplyChain_df, columns = ['Market', 'Order Country',
                                        'Order Region', 'Order State', 'Product Name', 
                                         'Category Name','Customer Segment'],drop_first=True, dtype=int)
```

```python
# Scaling

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaled_X=X.copy()
scaled_X=scaler.fit_transform(scaled_X)
scaled_X = pd.DataFrame(scaled_X, columns = X.columns)
```

# Model Building

# Random Forest Regressor

```python
# Get the train and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size = 0.20, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

rfr = RandomForestRegressor(random_state = 695)
forest_params = [{'n_estimators': [30, 50, 70, 90],
                  'max_features': [5, 10, 15, 20, 'sqrt', 'log2'],
                  'min_samples_split': [1, 50, 100, 150, 200]}]
```

```python
reg = GridSearchCV(rfr, forest_params, cv = 3, scoring='neg_mean_squared_error', verbose = True)
reg.fit(X_train, y_train)
```

```python
print(reg.best_params_)
print(reg.best_score_) # MSE
```

```python
# Test set
from sklearn.metrics import r2_score
y_pred = reg.predict(X_test)
mean_squared_error(y_test, y_pred)
```

```python
# Best model rerun
from sklearn.metrics import r2_score
rfr = RandomForestRegressor(random_state = 695, 
                            max_features = 20, min_samples_split = 50, n_estimators = 90)
rfr.fit(X_train, y_train)
y_pred_rf = rfr.predict(X_test)
print("R^2:", r2_score(y_test, y_pred_rf))
print("MSE:", mean_squared_error(y_test, y_pred_rf))
print("MAE:", mean_absolute_error(y_test, y_pred_rf))
```

```python
import pickle 

with open("randomforest_model.pkl","wb") as file:
    pickle.dump(rfr,file)
```

# Gradient Boosting Algorithm

```python
from sklearn.ensemble import GradientBoostingRegressor

gbt = GradientBoostingRegressor(random_state = 695)
boost_params = [{'n_estimators': [30, 50, 70, 90],
                  'max_features': [5, 10, 15, 20, 'sqrt', 'log2'],
                  'min_samples_split': [1, 50, 100, 150, 200]}]
```

```python
reg = GridSearchCV(gbt, boost_params, cv = 3, scoring='neg_mean_squared_error', verbose = 2)
reg.fit(X_train, y_train)
```

```python
# Convert the cv_results_ to a DataFrame
results_df = pd.DataFrame(reg.cv_results_)

# Select relevant columns to display
relevant_columns = ['rank_test_score', 'mean_test_score', 'std_test_score', 'params']

# Display the DataFrame sorted by rank_test_score
display_df = results_df.loc[:, relevant_columns].sort_values('rank_test_score')
display_df
```

```python
print(reg.best_params_)
print(reg.best_score_) # MSE
```

```python
# Test set
y_pred = reg.predict(X_test)
mean_squared_error(y_test, y_pred)
```
